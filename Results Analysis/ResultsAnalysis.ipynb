```python
def standardize_column(df, col):
    df.loc[:,col] = (df[col] - df[col].mean())/df[col].std()
```

```python
stats = ["coef","tstat","pval"]
```

```python
# Regression Analyses

dat = pd.DataFrame(index=pd.MultiIndex.from_product([trains, smalls,acs,fcs,pcs, batch_nums,endogs],names = ["train","small","ac","fc","pc","batch_num","endog"]), columns=[" ".join(x) for x in list(itertools.product(exogs,stats))])

for (train, small, ac, fc, pc) in itertools.product(trains, smalls,acs,fcs,pcs):
    
    df_all = pd.read_csv("../../Results/TestBig/reward_exp_dat__" + train + str(small) + str(ac) + str(fc) + str(pc) + ".csv",index_col="Unnamed: 0")
    df_all["sim_num"] = df_all.index

    for batch_num in batch_nums:
#         batch_num = 0 
# 
        df = df_all[df_all["batch_num"] == batch_num]
        df = df.sort_values(by="Mean", ascending = False)
#         standardize_column(df, "Mean")
        for endog in endogs:
            for exog in exogs: 
                model = sm.OLS(endog=df[endog], exog=sm.add_constant(df[exog])).fit()
                model.params[1], model.tvalues[1], model.pvalues[1]
                dat.loc[train,small,ac,fc,pc,batch_num,endog].loc[dat.columns.str.contains(exog)] = [model.params[1], model.tvalues[1], model.pvalues[1]]
```

```python
dat.to_csv("reg_results.csv")
```
